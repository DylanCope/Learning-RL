# The Behaviour of Stochastic Agents 

## Probabilty of Inequalities of Random Variables

When assessing the behaviour of a stochastic agent we look at the probability of one random variable taking a greater probability than another. To understand this, let's first consider the simplest situation of sampling two real numbers $X_i, X_j$ from a uniform distribution $\mathcal{U}$ and finding the probability that one is greater than the other $P(X_i > X_j)$. Intuitively, we understand that two independently and identically distributed variables have equal probability of falling either side of each other, so therefore $P(X_i > X_j) = \frac{1}{2}$, however, let's address this in a more formal proof,

\begin{align*}
    P(X_i > X_j) &= \int_{a}^{b}{\int_{a}^{X_i}{P(X_i) P(X_j)~dX_j}dX_i} \tag*{Given $X_i, X_j \sim \mathcal{U}(a, b)$}\\
    &= \int_{a}^{b}{\int_{a}^{X_i}{\bigg(\frac{1}{b - a}\bigg)^2~dX_j}dX_i} \tag*{As $P(X_i) = P(X_j) = \frac{1}{b-a}$}\\
    &= \bigg(\frac{1}{b - a}\bigg)^2\int_{a}^{b}{(X_i - a)~dX_i} \\
    &= \bigg(\frac{1}{b - a}\bigg)^2\bigg[\frac{1}{2}X_i^2 - ax_i\bigg]_{X_i=a}^{X_i = b} \\
    &= \bigg(\frac{1}{b - a}\bigg)^2\bigg(\frac{1}{2}b^2 - ab - \frac{1}{2}a^2 + a^2\bigg) \\
    &= \frac{1}{2}\bigg(\frac{1}{b - a}\bigg)^2\big(b^2 - 2ab + a^2\big) \\
    &= \frac{1}{2}\bigg(\frac{1}{b - a}\bigg)^2\big(b - a\big)^2 \\
    &= \frac{1}{2}
\end{align*}

Qualitatively, this approach sums the probabilities of every value of the variable on the right-hand side of the inequality where the value is less than some fixed value for the left-hand side. Thereby we can construct a more general for of the expression for non-identically distributed, non-independent random variables:

$$
    P(X_i > X_j) = \int^{S}{P(X_i \cap X_j)~dX_i dX_j}
$$

Where $S = \{~(x_i, x_j)~:~\forall~x_i \in U_i,~\forall~x_j < x_i \in U_j~\} $ and $U_i$ and $U_j$ are the sets from which $X_i$ and $X_j$ are sampled from. 

## Uniform Agents

For the sakes of our simple stochastic agent, we parameterize each random variable with the $Q_i$ and $Q_j$ 

\begin{align*}
    P(X_i) = Q_i X_i' \quad P(X_j) = Q_j X_j' \quad \text{where}~X_i', X_j' \sim \mathcal{U}(a, b)
\end{align*}

Hence, the random variables $X_i, X_j$ are independent and non-identically distributed.

\begin{align*}
    P(X_i > X_j) &= \int^{S}{P(X_i)P(X_j)~dX_i dX_j} \\
    &= \int_{aQ_i}^{bQ_i}{\int_{aQ_j}^{X_i}{P(X_i) P(X_j)~dX_j}dX_i} \\
    &= \int_{aQ_i}^{bQ_i}{\int_{aQ_j}^{X_i}{\frac{1}{Q_iQ_j}\bigg(\frac{1}{b-a}\bigg)^2~dX_j}dX_i} \\
    &= \frac{1}{Q_iQ_j}\bigg(\frac{1}{b-a}\bigg)^2\int_{aQ_i}^{bQ_i}{(X_i - aQ_j)~dX_i} \\
    &= \frac{1}{Q_iQ_j}\bigg(\frac{1}{b-a}\bigg)^2 \bigg[\frac{1}{2}X_i^2 - aQ_jX_i\bigg]_{X_i = aQ_i}^{X_i = bQ_i} \\
    &= \frac{1}{Q_iQ_j}\bigg(\frac{1}{b-a}\bigg)^2 \bigg(\frac{1}{2}b^2Q_i^2 - abQ_jQ_i - \frac{1}{2}a^2Q_i^2 + a^2Q_iQ_j\bigg)\\
    &= \frac{1}{2Q_iQ_j}\bigg(\frac{1}{b-a}\bigg)^2 \bigg(b^2Q_i^2 - 2abQ_jQ_i + a^2\big(2Q_iQ_j - Q_i^2\big)\bigg)\\
    &= \frac{1}{2}\bigg(\frac{1}{b-a}\bigg)^2 \bigg(b^2\frac{Q_i}{Q_j} - 2ab + a^2\big(2 - \frac{Q_i}{Q_j}\big)\bigg)
\end{align*}

Let $a = -z,~b = 1$

\begin{align*}
    \therefore~P(X_i > X_j) &= \frac{1}{2}\bigg(\frac{1}{z+1}\bigg)^2 \bigg(\frac{Q_i}{Q_j} + 2z + z^2\big(2 - \frac{Q_i}{Q_j}\big)\bigg) \\
    &= \frac{1}{2}\bigg(\frac{1}{z+1}\bigg)^2 \bigg(2z(z+1) - \frac{Q_i}{Q_j}(z^2 - 1)\bigg) \\
    &= \frac{1}{2}\bigg(\frac{1}{z+1}\bigg)^2 \bigg(2z(z+1) - \frac{Q_i}{Q_j}(z + 1)(z - 1)\bigg) \\
    &= \frac{1}{2}\bigg(\frac{1}{z+1}\bigg) \bigg(2z - \frac{Q_i}{Q_j}(z - 1)\bigg) \\
    &= \frac{z(2-r) + r}{2(z+1)} \tag*{Substituting $r = \frac{Q_i}{Q_j}$} \\
\end{align*}

Thereby the probability function is a linear in the ratio between $Q_i$ and $Q_j$,

$$
    P(X_i > X_j) = r\bigg(\frac{1 - z}{2 + 2z}\bigg) + \frac{z}{1 + z}
$$

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

r = np.linspace(-10, 10, 1000)
def p(r, z):
    return np.true_divide(z*(2 - r) + r, 2*z + 2)

def plot(explore, zoom):
    fig = plt.figure(figsize = (15, 9))
    ax = fig.add_subplot(111)
    y = p(r, explore)
    ax.plot(r, y, label='s_i over s_j')
    ax.plot(r, 1 - y, label='s_j over s_i')
    ax.set_xlim([1, zoom])
    ax.set_ylim([0, 1])
    ax.set_xlabel('Ratio between Q_i and Q_j')
    ax.set_ylabel('Probability')
    ax.legend()
    ax.grid()
    plt.show()

interact(plot, explore=(-0.95, 10.0, 0.05), zoom=(0.5, 15, 0.1))
```

As seen above, the use of an uniform distribution has the problem whereby after one action has a particularly greater $Q$-value than another it becomes impossible for any further exploration as the probability of choosing $s_i$ over $s_j$ reaches one. This being due to the fact that there exist zero-probability values in the uniform distribution. 

## Gaussian Agents

To alleviate this, let's sample according to Guassian distributions, 

$$
    X_i \sim \mathcal{N}\big(Q_i,~ z^2\big) \quad \quad
    X_j \sim \mathcal{N}\big(Q_j,~ z^2\big)
$$

Recall the Gaussian probability density function (PDF),

$$
    {\displaystyle f(x\;|\;\mu ,\sigma ^{2})={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}\;\exp\bigg(-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}\bigg)}}
$$

Which can also be expressed as a function of the standard normal distribution $\varphi$

$$
    f(x~|~\mu, \sigma^2) = \frac{1}{\sigma} \varphi\bigg(\frac{x - \mu}{\sigma}\bigg)
$$

Hence the inequality probability is given by

\begin{align*}
    P(X_i > X_j) = \int_{-\infty}^{\infty}{\frac{1}{z} \varphi\bigg(\frac{X_i - Q_i}{z}\bigg)\int_{-\infty}^{X_i}{\frac{1}{z} \varphi\bigg(\frac{X_j - Q_j}{z}\bigg)~dX_j}dX_i} \\
\end{align*}

Notice that the inner integral is the cumulative distribution function (CDF) of the distribution $P(X_i)$, where we are accumulating until the variable $X_j$. For any arbitrary gaussian distribution, the CDF $F$ can be expressed as a function of the CDF of $\varphi$, which is denoted $\Phi$.

\begin{align*}
    \Phi(x) &= \int_{-\infty}^{x} \varphi(t)~dt \\
    \therefore~F_{X_j}(x) &= \Phi\bigg(\frac{x - Q_j}{z}\bigg) \\
\end{align*}
Substituting this into the integral,
\begin{align*}
    P(X_i > X_j) &= \int_{-\infty}^{\infty} \frac{1}{z}\phi\bigg(\frac{X_i - Q_i}{z}\bigg)~\Phi\bigg(\frac{X_i - Q_j}{z}\bigg)~dX_i
\end{align*}
Which is simple enough to be solved analytically:

```python
from scipy.stats import norm
from scipy import integrate

def integrand(x, z=1, q_i=0, q_j=1):
    phi = norm.pdf((x - q_i) / z)
    Phi = norm.cdf((x - q_j) / z)
    return phi * Phi / z

def plot(z, q_i, q_j):
    x = np.linspace(-20, 20, 1000)
    plt.figure(figsize = (12, 8))
    i = integrand(x, z=z, q_i=q_i, q_j=q_j)
    plt.plot(x, i)
    plt.fill_between(x, i, label='$P(X_i > X_j)$', color='#DDDDEE')
    plt.plot(x, norm.pdf(x, loc=q_i, scale=z), label='$P(X_i)$')
    plt.plot(x, norm.pdf(x, loc=q_j, scale=z), label='$P(X_j)$')
    dx = (x.max() - x.min()) / x.size
    prob_est = i.sum() * dx
    plt.annotate('%.2f' % prob_est,
                 xy = (x[i.argmax()], i.max() / 2.0),
                 horizontalalignment = 'center')
    plt.grid()
    plt.legend(prop = {'size' : 17}, loc='upper left', bbox_to_anchor=(1, 0.95))
    plt.show()
    
_ = interact(plot, z = (0.1, 10.), q_i = (-10., 10.), q_j = (-10., 10.))
```

This interactive graph shows how varying the parameters changes the likelihood of outcomes.

```python
@np.vectorize
def prob(x, q_i=0, q_j=0):
    f = lambda y : integrand(y, z=x, q_i=q_i, q_j=q_j)
    p, err = integrate.quad(f, -np.inf, np.inf)
    return p

x = np.linspace(0.05, 5, 50)
plt.figure(figsize = (12, 8))
for q_i in [-4, -2, -1, -0.5, -0.25, 0, 0.25, 0.5, 1, 2, 4]:
    plt.plot(x, prob(x, q_i=q_i), label = '$Q_i = %.2f$' % q_i)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), prop = {'size' : 15})
plt.title('$P(X_i > X_j)$ with fixed $Q_j = 0$ and varying $Q_i$', fontsize=20)
plt.xlabel('Exploration parameter', fontsize=20)
plt.ylabel('$P(X_i > X_j)$', fontsize=20)
plt.show()
```

Here we see that for states with increasingly disparate $Q$-values it becomes increasingly likely that the state with the large associated $Q$ is chosen.
